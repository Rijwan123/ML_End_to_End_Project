# retrieve the row from the DataFrame df where the value in the 'first_ings_score' column is equal to the ***maximum*** value present in that column

```python
# df.loc[df['first_ings_score'].idxmax()] in pandas is used to retrieve the row
# from the DataFrame df where the value in the 'first_ings_score' column is equal
# to the maximum value present in that column

# df.loc[]: This method is used for label-based indexing,
# allowing you to access a group of rows and columns by label(s) or a boolean array.

df.loc[df['first_ings_score'].idxmax()]
```

# retrieve the row from the DataFrame df where the value in the 'first_ings_score' column is equal to the **minimum** value present in that column

```python
df.loc[df['first_ings_score'].idxmin()]
```
